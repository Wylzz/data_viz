### Modèle linéaire_Calcul des coefficients de régression

Pour la suite de notre travail, et en nous basant sur nos dataviz, nous décidons de réaliser un modèle linéaire prenant en compte le score de récidive (decile_score.1) et les races des détenus.

```python
x = df_vr['race_num']
y = df_vr['decile_score.1']
```

```python
tmp=0
for i in range(len(x)):
    tmp=tmp+x[i]*y[i]
print(tmp,(df_vr['race_num']*df_vr['decile_score.1']).sum())
```

```python
#calcul de a

un=len(x)*(df_vr['race_num']*df_vr['decile_score.1']).sum()

sumx=df_vr['race_num'].sum()
sumy=df_vr['decile_score.1'].sum()

deux=len(df_vr['race_num'])*((df_vr['race_num']*df_vr['race_num']).sum())

trois = sumx*sumx

a = (un-(sumx*sumy))/(deux-trois)
a
```

```python
#calcul de b

quatre = sumy*((df_vr['race_num']*df_vr['race_num']).sum())
cinq = sumx*(df_vr['race_num']*df_vr['decile_score.1']).sum()
six = len(df_vr['race_num'])*((df_vr['race_num']*df_vr['race_num']).sum())
sept = sumx*sumx

b=(quatre-cinq)/(six-sept)
b
```
